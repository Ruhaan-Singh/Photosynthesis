```python
train = pd.DataFrame()
val = pd.DataFrame()
for i in database['Dataset_name'].unique():
    x = database[database['Dataset_name'] == i]
    
    add_to_train = x.sample(frac=0.8,random_state=42)
    add_to_val = x.drop(add_to_train.index)
    
    train = pd.concat([train,add_to_train])
    val = pd.concat([val,add_to_val])
    
train = train.sample(frac=1).reset_index()
val = val.sample(frac=1).reset_index()
```

```python
X_train = np.stack(train['Spectra'].to_numpy())
y_train = train['Jmax25']

X_val = np.stack(val['Spectra'].to_numpy())
y_val = val['Jmax25']
```

```python
error_by_component = []
for i in range(1,len(y_train)+1):
    plsr = PLSRegression(n_components=i)
    plsr.fit(X_train,y_train)
    predicted = plsr.predict(X_val).squeeze()
    error_by_component.append(((y_val - predicted)**2).mean())
```
