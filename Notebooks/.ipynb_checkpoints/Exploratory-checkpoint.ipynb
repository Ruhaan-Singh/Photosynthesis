```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
from sklearn.model_selection import LeaveOneOut
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import r2_score
import geopandas as gpd
```

```python
database1 = pd.read_csv("/Users/rishi/Documents/ORNL Internships/25/Database_p1.csv", encoding='latin1')
database2 = pd.read_csv("/Users/rishi/Documents/ORNL Internships/25/Database_p2.csv", encoding='latin1')
database3 = pd.read_csv("/Users/rishi/Documents/ORNL Internships/25/Database_p3.csv", encoding='latin1')
```

```python
database = pd.concat([database1,database2,database3],ignore_index=True)

len(database['Dataset_name'].unique())
```

```python
database = pd.concat([database1,database2,database3],ignore_index=True)
database = database.dropna(subset=['Jmax25'])
```

```python
database = database[database['Spectra_trait_pairing'] == "Same"]
database = database.reset_index()
database['sqrt_Jmax25'] = database['Jmax25'] ** 0.5
```

```python
plt.hist(database['Jmax25'])
```

```python
plt.hist(database['sqrt_Jmax25'])
```

```python
spectra = []
for i in range(len(database)):
    temp = []
    for j in range(400,2401):
        temp.append(database['Wave_ ' + str(j)][i])
    spectra.append(temp)
database['Spectra'] = spectra
database = database[['SampleID','Dataset_name','sqrt_Jmax25','Jmax25','StdError_Jmax25','Spectra']]
```

```python
dropped_rows = []
for i in range(len(database)):
    for j in range(len(database['Spectra'][i])):
        if np.isnan(database['Spectra'][i][j]):
                   dropped_rows.append(i)
database = database.drop(dropped_rows).reset_index()
```

```python
for i in range(len(database)):
    plt.plot(database['Spectra'][i])
```

```python
X = np.stack(database['Spectra'].to_numpy())
print(np.shape(X))
y = np.array(database['Jmax25'])
print(np.shape(y))
```

```python
predictions = []
Loo = LeaveOneOut()
for i, (train_index,val_index) in enumerate(Loo.split(X)):
    X_train = X[train_index]
    y_train = y[train_index]
    X_val = X[val_index]
    y_val = y[val_index]
    plsr = PLSRegression(n_components=19)
    plsr.fit(X_train,y_train)
    
    predictions.append(plsr.predict(X_val))
    
```

```python
diff = []

for i in range(len(y)):
    diff.append(y[i] - predictions[i])
diff = np.array(diff).squeeze()
```

```python
outliers_indices = []
above_threshold = diff.mean() + 2.6 * diff.std()
below_threshold = diff.mean() - 2.6 * diff.std()

for i in range(len(diff)):
    if (diff[i] < below_threshold) or (diff[i] > above_threshold):
        outliers_indices.append(i)
        
plt.plot(diff,color='r')
plt.axhline(y=above_threshold)
plt.axhline(y=below_threshold)
```

```python
database = database.drop(outliers_indices).reset_index()
database = database[['SampleID','Dataset_name','sqrt_Jmax25','Jmax25','StdError_Jmax25','Spectra']]
```

```python
train = pd.DataFrame()
val = pd.DataFrame()
for i in database['Dataset_name'].unique():
    x = database[database['Dataset_name'] == i]
    
    add_to_train = x.sample(frac=0.8,random_state=42)
    add_to_val = x.drop(add_to_train.index)
    
    train = pd.concat([train,add_to_train])
    val = pd.concat([val,add_to_val])
    
train = train.sample(frac=1).reset_index()
val = val.sample(frac=1).reset_index()
```

```python
X_train = np.stack(train['Spectra'].to_numpy())
y_train = train['Jmax25']

X_val = np.stack(val['Spectra'].to_numpy())
y_val = val['Jmax25']
```

```python
print(max(y_val) - min(y_val))
```

```python
error_by_component = []
for i in range(1,201):
    plsr = PLSRegression(n_components=i)
    plsr.fit(X_train,y_train)
    predicted = plsr.predict(X_val).squeeze()
    error_by_component.append(((y_val - predicted)**2).mean())
```

```python
plt.plot(error_by_component,color='r')
min_index = error_by_component.index(min(error_by_component))
print(min_index)
print(min(error_by_component))
plt.axvline(x=min_index)
```

```python
plsr = PLSRegression(n_components=39)
plsr.fit(X_train,y_train)
predicted = plsr.predict(X_val).squeeze()
sns.regplot(x=predicted,y=y_val,line_kws=dict(color='red'))
r2 = r2_score(predicted, y_val)
print(r2)
```

```python
database1 = pd.read_csv("/Users/rishi/Documents/ORNL Internships/25/Database_p1.csv", encoding='latin1')
database2 = pd.read_csv("/Users/rishi/Documents/ORNL Internships/25/Database_p2.csv", encoding='latin1')
database3 = pd.read_csv("/Users/rishi/Documents/ORNL Internships/25/Database_p3.csv", encoding='latin1')
database = pd.concat([database1,database2,database3],ignore_index=True)

biomes = pd.read_csv("/Users/rishi/Documents/ORNL Internships/25/Biomes.csv")
names = []
num_observations = []
num_species = []
for i in database['Biome_number'].unique():
    
    value = biomes.loc[biomes['Biome_number'] == i, 'Biome'].iloc[0]
    names.append(value)
    
    temp = database[database['Biome_number'] == i]
    num_observations.append(len(temp))
    num_species.append(len(temp['Species'].unique()))
    
    
```

```python
sorted_pairs = sorted(zip(num_observations, names))
sorted_values, sorted_categories = zip(*sorted_pairs)

plt.barh(sorted_categories,sorted_values)
plt.xlabel('Number of Observations')
```

```python
sorted_pairs = sorted(zip(num_species, names))
sorted_values, sorted_categories = zip(*sorted_pairs)

plt.barh(sorted_categories,sorted_values)
plt.xlabel('Number of Species')
```

```python
shape = gpd.read_file("/Users/rishi/Documents/ORNL Internships/25/official/wwf_terr_ecos.shp")
merged_gdf = shape.merge(biomes, left_on='BIOME',right_on='Biome_number', how='inner')
```

```python
man_dataset = []
man_lat = []
man_long = []

dataset = []
lat = []
long = []

for i in database['Latitude'].unique():
    
    temp = database[database['Latitude'] == i]
    temp.reset_index(inplace=True)
    if temp['Biome_number'][0] < 15:
        dataset.append(i)
        lat.append(temp['Latitude'][0])
        long.append(temp['Longitude'][0])
    else:
        man_dataset.append(i)
        man_lat.append(temp['Latitude'][0])
        man_long.append(temp['Longitude'][0])        

man_points = pd.DataFrame({'Dataset':man_dataset,'Lat':man_lat,'Long':man_long})

man_dataset_loc = gpd.GeoDataFrame(
    man_points, geometry=gpd.points_from_xy(man_points.Long, man_points.Lat), crs="EPSG:4326"
)
        
points = pd.DataFrame({'Dataset':dataset,'Lat':lat,'Long':long})

dataset_loc = gpd.GeoDataFrame(
    points, geometry=gpd.points_from_xy(points.Long, points.Lat), crs="EPSG:4326"
)
```

```python
import matplotlib.pyplot as plt

# Plot the base layer with biome coloring
base = merged_gdf.plot(column='Biome', legend=True, figsize=(10, 8))

# Overlay dataset points with labels for legend
man_dataset_loc.plot(ax=base, color='black', marker='o', facecolor='none', edgecolor='black', label='Managed')
dataset_loc.plot(ax=base, color='black', marker='^', facecolor='none', edgecolor='black', label='Natural')

# Manually move the combined legend (from points) outside the plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Point Sources')

# Tidy up layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import Patch

unique_biomes = merged_gdf['Biome'].unique()
colors = plt.cm.Set3(range(len(unique_biomes)))  # or another colormap

biome_color_map = dict(zip(unique_biomes, colors))

fig, ax = plt.subplots(figsize=(10, 8))
for biome, color in biome_color_map.items():
    merged_gdf[merged_gdf['Biome'] == biome].plot(ax=ax, color=color, label=biome)

man_dataset_loc.plot(ax=ax, color='black', marker='o', facecolor='none', edgecolor='black')
dataset_loc.plot(ax=ax, color='black', marker='^', facecolor='none', edgecolor='black')

biome_patches = [Patch(facecolor=color, edgecolor='black', label=biome) for biome, color in biome_color_map.items()]
point_handles = [
    Line2D([0], [0], marker='o', color='black', label='Manual Dataset',
           markerfacecolor='none', markeredgecolor='black', linestyle='None'),
    Line2D([0], [0], marker='^', color='black', label='Dataset',
           markerfacecolor='none', markeredgecolor='black', linestyle='None')
]

all_handles = biome_patches + point_handles

ax.legend(handles=all_handles, loc='center left', bbox_to_anchor=(1, 0.5), title='Legend')

plt.tight_layout()
plt.show()
```
